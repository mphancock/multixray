```python
from pathlib import Path 
import numpy as np 
import pandas as pd
```

```python
sample_score_dir = Path(Path.home(), "Documents/xray/sample_bench/dev/01_w_xray/data/sample_scores")
analysis_dict = dict()
analysis_dict["job_num"]
analysis_dict["rmsd_min"]
analysis_dict["rmsd_mu"]
analysis_dict["rmsd_sig"]
analysis_dict["rmsd_align_min"]
analysis_dict["rmsd_align_mu"]
analysis_dict["rmsd_align_sig"]


for sample_score_file in sample_score_dir.glob("*"):
    print(sample_score_file)
    job_num = int(Path(sample_score_file).stem)
    
    score_df = pd.read_csv(sample_score_file, index_col=0)
    min_rmsd = np.min(score_df["rmsd"])
    min_rmsd_align = np.min(score_df["rmsd_align"])
    
    print(job_num, min_rmsd, min_rmsd_align)

```
