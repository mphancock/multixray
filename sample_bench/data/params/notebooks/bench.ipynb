```python
w_xray_df = pd.read_csv(Path(Path.home(), "Documents/xray/sample_bench/data/7mhf/186_w_xray_bench_ref_10000/best_w_xray.csv"), index_col=0)
w_xray_df.head()
```

```python
w_xray_df_tmp = w_xray_df.copy()
w_xray_df_tmp.fillna(0, inplace=True)
cif_df_tmp = cif_df.copy()
cif_df_tmp.fillna(0, inplace=True)
```

```python
input_df = pd.DataFrame()

input_id = 0
for N in [1, 2]:
    for j in range(20):
        input_df.loc[input_id, "N"] = N
        if j % 2 == 0:
            input_df.loc[input_id, "J"] = 1
        else:
            input_df.loc[input_id, "J"] = 2

        for col in ["cif_0", "cif_1", "ref_0", "ref_1", "w_0_0", "w_1_0", "w_0_1", "w_1_1"]:
            input_df.loc[input_id, col] = cif_df.loc[j, col]

        # print(cif_df.loc[j, "cif_0"], cif_df.loc[j, "cif_1"])
        w_xray_row = w_xray_df.loc[(w_xray_df_tmp["cif_0"] == cif_df_tmp.loc[j, "cif_0"]) & (w_xray_df_tmp["cif_1"] == cif_df_tmp.loc[j, "cif_1"]) & (w_xray_df_tmp["N"] == N)]
        w_xray = w_xray_row["w_xray"].values[0]
        # w_xray = w_xray_df.loc[input_id, "w_xray"]
        input_df.loc[input_id, "w_xray"] = w_xray

        input_id += 1

input_df["sample_sched_str"] = "step3000,T300,dofA,pdb1,w1,res2"
input_df["start_pdb_file"] = "/wynton/home/sali/mhancock/xray/data/pdbs/7mhf/7mhf_refine.pdb"

# input_df["N"] = input_df["N"].astype(int)

input_df.head()
```
