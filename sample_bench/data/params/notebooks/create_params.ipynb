```python
param_df = pd.DataFrame()
job_id = 0

for start_id in range(10):
    start_pdb_file = "/wynton/home/sali/mhancock/xray/dev/40_higher_energy/data/37/{}.pdb".format(start_id)

    for w_xray in w_xrays:
        param_df.loc[job_id, "start_pdb_file"] = start_pdb_file

        param_df.loc[job_id, "N"] = 1
        param_df.loc[job_id, "J"] = 1
        param_df.loc[job_id, "cif_0"] = Path("/wynton/home/sali/mhancock/xray/dev/39_bench_ensemble/data/cifs/7mhl.cif")
        param_df.loc[job_id, "ref_0"] = Path("/wynton/home/sali/mhancock/xray/dev/39_bench_ensemble/data/pdbs/7mhl.pdb")
        param_df.loc[job_id, "w_xray"] = w_xray
        job_id += 1

param_df["sample_sched_str"] = "step3000,T300,dofA,pdb1,w0,res0"
param_df["init_weights"] = "uni"

param_df.head()
```
