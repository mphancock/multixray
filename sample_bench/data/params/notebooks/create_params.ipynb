```python
param_df = pd.DataFrame()

cif_dir = Path("/wynton/home/sali/mhancock/xray/dev/38_standard_flags/data/")
pdb_files = list()
for pdb_name in ["7mhf", "7mhg", "7mhh", "7mhi", "7mhj", "7mhk"]:
    pdb_files.append(Path(cif_dir, "{}.cif".format(pdb_name)))

pdb_combos = list(combinations(pdb_files, 3))
job_id = 0
for pdb_combo in pdb_combos:
    for j in range(len(pdb_combo)):
        param_df.loc[job_id, "cif_{}".format(j)] = pdb_combo[j]
    job_id += 1

param_df["init_pdbs"] = Path("/wynton/home/sali/mhancock/xray/data/pdbs/7mhf/7mhf_refine.pdb")
param_df["N"] = 2
param_df["ref_N"] = 1
param_df["J"] = 3

param_df["ref_pdbs"] = Path("/wynton/home/sali/mhancock/xray/data/pdbs/7mhf/7mhf_refine.pdb")
for j in range(len(pdb_combo)):
    param_df["ref_0_{}".format(j)] = 1

param_df["w_xray"] = 0.5
param_df["xray_freq"] = 1
param_df["weight_thermo"] = 0
param_df["vel_thermo"] = 1
param_df["refine"] = 1
param_df["sample_sched_str"] = "s5000,t5000,dA,p1,w0,r0"

print(len(param_df))
param_df.head()
```
