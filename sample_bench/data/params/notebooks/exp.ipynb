```python
cif_df = pd.read_csv(Path(Path.home(), "Documents/xray/dev/35_cif_combos/data/7mhf.csv"), index_col=0)
cif_df.head()
```

```python
w_xray_df = pd.read_csv(Path(Path.home(), "Documents/xray/sample_bench/data/7mhf/188_w_xray_exp_ref_10000/best_w_xray.csv"), index_col=0)
w_xray_df.head()
```

```python
w_xray_df_tmp = w_xray_df.copy()
w_xray_df_tmp.fillna(0, inplace=True)
cif_df_tmp = cif_df.copy()
cif_df_tmp.fillna(0, inplace=True)
```

```python
cif_cols = ["cif_{}".format(i) for i in range(6)]
ref_cols = ["ref_{}".format(i) for i in range(6)]
```

```python
len(cif_df.loc[6, cif_cols].dropna())
```

```python
input_df = pd.DataFrame()

input_id = 0
for N in [1, 2, 4, 8, 16]:
    for j in range(63):
        input_df.loc[input_id, "N"] = N
        input_df.loc[input_id, "J"] = len(cif_df.loc[j, cif_cols].dropna())

        for i in range(len(cif_cols)):
            input_df.loc[input_id, "cif_{}".format(i)] = cif_df.loc[j, "cif_{}".format(i)]
            input_df.loc[input_id, "ref_{}".format(i)] = cif_df.loc[j, "ref_{}".format(i)]
            input_df.loc[input_id, "w_{}_{}".format(0, i)] = 1

        w_xray_row = w_xray_df.loc[(w_xray_df_tmp["cif_0"] == cif_df_tmp.loc[j, "cif_0"]) & (w_xray_df_tmp["cif_1"] == cif_df_tmp.loc[j, "cif_1"]) & (w_xray_df_tmp["cif_2"] == cif_df_tmp.loc[j, "cif_2"]) & (w_xray_df_tmp["cif_3"] == cif_df_tmp.loc[j, "cif_3"]) & (w_xray_df_tmp["cif_4"] == cif_df_tmp.loc[j, "cif_4"]) & (w_xray_df_tmp["cif_5"] == cif_df_tmp.loc[j, "cif_5"]) & (w_xray_df_tmp["N"] == N)]
        w_xray = w_xray_row["w_xray"].values[0]
        input_df.loc[input_id, "w_xray"] = w_xray

        input_id += 1

input_df["sample_sched_str"] = "step3000,T300,dofA,pdb1,w1,res0"
input_df["start_pdb_file"] = "/wynton/home/sali/mhancock/xray/data/pdbs/7mhf/7mhf_refine.pdb"

# input_df["N"] = input_df["N"].astype(int)

input_df.head()
```

```python
input_df.to_csv(Path(Path.home(), "Documents/xray/sample_bench/data/params/exp.csv"))
```
