```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python

Ns = [1, 2, 4, 8, 16]
w_xrays = [2**i for i in range(-5,6)]
w_xrays.sort()
```

```python
cif_df = pd.read_csv(Path(Path.home(), "Documents/xray/dev/35_cif_combos/data/7mhf.csv"), index_col=0)
job_ids = cif_df.index
cif_df.head()
```

```python
input_df = pd.DataFrame()

for N in Ns:
    for job_id in job_ids:
        for w_xray in w_xrays:
            input_id = len(input_df)
            input_df.loc[input_id, "N"] = N

            cif_cols = [col for col in cif_df.columns if "cif" in col]
            cifs = cif_df.loc[job_id, cif_cols]

            J = cifs.notna().sum()

            input_df.loc[input_id, "J"] = J
            for col in cif_df.columns:
                input_df.loc[input_id, col] = cif_df.loc[job_id, col]

            input_df.loc[input_id, "w_xray"] = w_xray

input_df["sample_sched_str"] = "step3000,T300,dofA,pdb1,w1,res0"
input_df["start_pdb_file"] = "/wynton/home/sali/mhancock/xray/data/pdbs/7mhf/7mhf_refine.pdb"

# input_df["N"] = input_df["N"].astype(int)

input_df.head()
```

```python
input_df.to_csv(Path(Path.home(), "Documents/xray/sample_bench/data/params/w_xray_exp.csv"))
```
