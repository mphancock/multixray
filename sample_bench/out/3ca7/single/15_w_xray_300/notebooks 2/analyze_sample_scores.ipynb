```python
from pathlib import Path 
import numpy as np 
import pandas as pd
```

```python
sample_score_dir = Path(Path.home(), "Documents/xray/sample_bench/dev/01_w_xray/data/sample_scores")
analysis_dict = dict()
analysis_dict["job_num"] = list() 
analysis_dict["rmsd_min"] = list() 
analysis_dict["rmsd_mu"] = list() 
analysis_dict["rmsd_sig"] = list() 
analysis_dict["rmsd_align_min"] = list() 
analysis_dict["rmsd_align_mu"] = list() 
analysis_dict["rmsd_align_sig"] = list() 
analysis_dict["xray_min"] = list() 
analysis_dict["xray_mu"] = list() 
analysis_dict["xray_sig"] = list() 


for sample_score_file in sample_score_dir.glob("*"):
    print(sample_score_file)
    job_num = int(Path(sample_score_file).stem)
    
    score_df = pd.read_csv(sample_score_file, index_col=0)
    analysis_dict["job_num"].append(job_num)
    analysis_dict["rmsd_min"].append(np.min(score_df["rmsd"]))
    analysis_dict["rmsd_mu"].append(np.mean(score_df["rmsd"]))
    analysis_dict["rmsd_sig"].append(np.std(score_df["rmsd"]))
    analysis_dict["rmsd_align_min"].append(np.min(score_df["rmsd_align"]))
    analysis_dict["rmsd_align_mu"].append(np.mean(score_df["rmsd_align"]))
    analysis_dict["rmsd_align_sig"].append(np.std(score_df["rmsd_align"]))
    analysis_dict["xray_min"].append(np.min(score_df["xray_score"])) 
    analysis_dict["xray_mu"].append(np.mean(score_df["xray_score"]))
    analysis_dict["xray_sig"].append(np.std(score_df["xray_score"]))
    print(job_num)

analysis_df = pd.DataFrame(analysis_dict)
analysis_df.to_csv(Path(Path.home(), "Documents/xray/sample_bench/dev/01_w_xray/data/analysis.csv"))
```
