```python
from pathlib import Path
import pandas as pd
```

```python
cif_df = pd.read_csv(Path(Path.home(), "Documents/xray/dev/29_synthetic_native_3/data/cifs/csvs/7mhf_30.csv"), index_col=0)
job_ids = cif_df.index
Ns = [1, 2]
w_xrays = [2**i for i in range(-5,6)]
w_xrays.sort()
```

```python
input_df = pd.DataFrame()

for N in Ns:
    for j in range(20):
        cif_0 = cif_df.loc[j, "cif_0"]
        cif_1 = cif_df.loc[j, "cif_1"]
        ref_0 = cif_df.loc[j, "ref_0"]
        ref_1 = cif_df.loc[j, "ref_1"]
        w_0_0 = cif_df.loc[j, "w_0_0"]
        w_1_0 = cif_df.loc[j, "w_1_0"]
        w_0_1 = cif_df.loc[j, "w_0_1"]
        w_1_1 = cif_df.loc[j, "w_1_1"]

        for w_xray in w_xrays:
            input_id = len(input_df)

            input_df.loc[input_id, "N"] = N
            if j % 2 == 0:
                input_df.loc[input_id, "J"] = 1
            else:
                input_df.loc[input_id, "J"] = 2

            input_df.loc[input_id, "cif_0"] = cif_0
            input_df.loc[input_id, "cif_1"] = cif_1
            input_df.loc[input_id, "ref_0"] = ref_0
            input_df.loc[input_id, "ref_1"] = ref_1
            input_df.loc[input_id, "w_0_0"] = w_0_0
            input_df.loc[input_id, "w_1_0"] = w_1_0
            input_df.loc[input_id, "w_0_1"] = w_0_1
            input_df.loc[input_id, "w_1_1"] = w_1_1

            input_df.loc[input_id, "w_xray"] = w_xray

input_df["sample_sched_str"] = "step3000,T300,dofA,pdb1,w1,res2"
input_df["start_pdb_file"] = "/wynton/home/sali/mhancock/xray/data/pdbs/7mhf/7mhf_refine.pdb"

# input_df["N"] = input_df["N"].astype(int)

input_df.head()
```

```python
input_df.to_csv(Path(Path.home(), "Documents/xray/sample_bench/data/params/w_xray_bench.csv"))
```
