```python
from pathlib import Path
import pandas as pd

import IMP
import IMP.atom
```

```python
sel = IMP.atom.AndPDBSelector(IMP.atom.NonWaterNonHydrogenPDBSelector(), IMP.atom.NonAlternativePDBSelector())
pdb_file = Path(Path.home(), "Documents/xray/sample_bench/data/7mhf/166_N1/summary_15/7mhh_N8.pdb")
m = IMP.Model()
hs = IMP.atom.read_multimodel_pdb(str(pdb_file), m, sel)

ref_pdb_file = Path(Path.home(), "Documents/xray/data/pdbs/7mhf/7mhh.pdb")
ref_m = IMP.Model()
ref_h = IMP.atom.read_pdb(str(ref_pdb_file), ref_m, sel)
```

```python
rmsd_df = pd.DataFrame(index=list(range(1,307)),columns=["rmsd", "aa"])
rmsd_df = rmsd_df.fillna(0)
for res in range(1,307):

    ref_sel = IMP.atom.Selection(ref_h, residue_index=res)
    ref_pids = ref_sel.get_selected_particle_indexes()
    ref_xyzs = [IMP.core.XYZ(ref_m, pid) for pid in ref_pids]

    at = IMP.atom.Atom(ref_m, ref_pids[0])
    rmsd_df.loc[res, "aa"] = str(IMP.atom.get_residue(at).get_residue_type())

    for h in hs:
        sel = IMP.atom.Selection(h, residue_index=res)
        pids = sel.get_selected_particle_indexes()
        xyzs = [IMP.core.XYZ(m, pid) for pid in pids]

        # print(res, IMP.atom.get_rmsd(ref_xyzs, xyzs))
        rmsd_df.loc[res, "rmsd"] = rmsd_df.loc[res, "rmsd"] + IMP.atom.get_rmsd(ref_xyzs, xyzs)
        # rmsd_df.loc[res, "rmsd"] = 1

    # print(len(pids), len(ref_pids))
```
