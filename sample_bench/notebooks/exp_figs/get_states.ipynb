```python
from pathlib import Path

import IMP
import IMP.atom
```

```python
data_dir = Path(Path.home(), "Documents/xray/sample_bench/data/7mhf/189_exp_ref_10000")
pdb_dir = Path(data_dir, "sample")
states_pdb_dir = Path(data_dir, "summary_states")
for pdb_file in pdb_dir.glob("*"):
    print(pdb_file)

    m = IMP.Model()
    hs = IMP.atom.read_multimodel_pdb(str(pdb_file), m, IMP.atom.AllPDBSelector())

    for i in range(len(hs)):
        out_pdb_file = Path(states_pdb_dir, "{}_{}.pdb".format(pdb_file.stem, i))
        IMP.atom.write_pdb(hs[i], str(out_pdb_file))
```
