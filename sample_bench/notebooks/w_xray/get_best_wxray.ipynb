```python
import pandas as pd
from pathlib import Path
```

```python
exp_name = "188_w_xray_exp_ref_10000"
w_xray_df = pd.read_csv(Path(Path.home(), "Documents/xray/sample_bench/data/7mhf/{}/w_xray.csv".format(exp_name)), index_col=0)
w_xray_df.head()

w_xray_filt_df = w_xray_df.loc[w_xray_df["ff"] < 200000]
# w_xray_filt_df = w_xray_df
w_xray_filt_df.head()
```

```python
# wxray_filt_df = wxray_df

match_cols = ["N", "J"]
cif_cols = [col for col in w_xray_filt_df.columns if "cif" in col]
match_cols.extend(cif_cols)
print(match_cols)
# min_idx = w_xray_filt_df.groupby(match_cols)['r_free'].idxmin()
w_xray_filt_df_tmp = w_xray_filt_df.fillna('nan')
min_idx = w_xray_filt_df_tmp.groupby(match_cols)["r_free"].idxmin()

best_w_xray_df = w_xray_filt_df.loc[min_idx]
# best_wxray_df['w_xray'] = best_wxray_df['w_xray_id'].map(lambda x: w_xrays[int(x)])
# best_wxray_df["N"] = best_wxray_df['state_id'].map(lambda x: n_states[int(x)])

print(len(best_w_xray_df))
best_w_xray_df.reset_index(drop=True, inplace=True)
best_w_xray_df.head()
```
