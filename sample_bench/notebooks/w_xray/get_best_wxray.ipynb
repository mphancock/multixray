```python
import pandas as pd
from pathlib import Path
```

```python
exp_name = "180_wxray_bench"
wxray_df = pd.read_csv(Path(Path.home(), "Documents/xray/sample_bench/data/7mhf/{}/w_xray.csv".format(exp_name)), index_col=0)
wxray_df.head()
```

```python
w_xrays = [0.03125, 0.0625, .125, .25, .5, 1, 2, 4, 8, 16, 32]
n_states = [1, 2, 4, 8, 16]
job_ids = list(range(20))
```

```python
wxray_filt_df = wxray_df.loc[wxray_df["ff"] < 10000]
# wxray_filt_df = wxray_df

min_idx = wxray_filt_df.groupby(['state_id', 'job_id'])['field'].idxmin()

best_wxray_df = wxray_filt_df.loc[min_idx]
best_wxray_df['w_xray'] = best_wxray_df['w_xray_id'].map(lambda x: w_xrays[int(x)])
best_wxray_df["N"] = best_wxray_df['state_id'].map(lambda x: n_states[int(x)])

best_wxray_df.head()
```

```python
best_wxray_df.tail()
```

```python
best_wxray_df.to_csv(Path(Path.home(), "Documents/xray/sample_bench/data/7mhf/{}/best_wxray.csv".format(exp_name)))
```

```python
for state_id in range(len(n_states)):
    for job_id in job_ids:
        w_xray_N_J = wxray_df.loc[(wxray_df["state_id"] == state_id) & (wxray_df["job_id"] == job_id)]
        w_xray_N_J_filt = w_xray_N_J.loc[w_xray_N_J["ff"] < 100000]

        w_xray_N_J_min_ff = w_xray_N_J_filt.loc[w_xray_N_J_filt["field"].idxmin()]["ff"]
        w_xray_N_J_min_wxray = w_xrays[int(w_xray_N_J_filt.loc[w_xray_N_J_filt["field"].idxmin()]["w_xray_id"])]

        print(state_id, job_id, len(w_xray_N_J), len(w_xray_N_J_filt), w_xray_N_J_min_wxray)

```
