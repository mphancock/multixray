```python
import pandas as pd
from pathlib import Path
```

```python
wxray_df = pd.read_csv(Path(Path.home(), "Documents/xray/sample_bench/data/7mhf/178_wxray/w_xray.csv"), index_col=0)
```

```python
w_xrays = [0.03125, 0.0625, .125, .25, .5, 1, 2, 4, 8, 16, 32]
n_states = [1, 2, 4, 8, 16]
```

```python
min_idx = wxray_df.groupby(['state_id', 'job_id'])['field'].idxmin()

# Use the indices to filter the rows from the original dataframe
best_wxray_df = wxray_df.loc[min_idx]
best_wxray_df['w_xray'] = best_wxray_df['w_xray_id'].map(lambda x: w_xrays[int(x)])
best_wxray_df["N"] = best_wxray_df['state_id'].map(lambda x: n_states[int(x)])

best_wxray_df.head()
```
