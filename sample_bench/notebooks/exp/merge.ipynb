```python
from pathlib import Path
import pandas as pd
import random
import numpy as np
```

```python
data_dir = Path(Path.home(), "Documents/xray/sample_bench/data/7mhf/182_bench")

ref_dfs = list()
for i in range(45):
    ref_df_file = Path(data_dir, "ref_100000_{}.csv".format(i*100))
    ref_df = pd.read_csv(ref_df_file, index_col=0)
    ref_dfs.append(ref_df)

merge_df = pd.concat(ref_dfs)
merge_df.reset_index(drop=True, inplace=True)
merge_df.head()
```
