```python
score_dfs = np.ndarray(shape=(files_arr.shape[0], files_arr.shape[1]), dtype=object)
for i in range(files_arr.shape[0]):
    for j in range(files_arr.shape[1]):
        score_df = pd.read_csv(files_arr[i,j], index_col=0)

        if j == 1:
            score_df["rmsd_0+rmsd_1"] = (score_df["rmsd_0"] + score_df["rmsd_1"])/2

        score_dfs[i,j] = score_df
```

```python
for i in range(score_dfs.shape[0]):
    for j in range(score_dfs.shape[1]):
        print(i,j)
        for score_field in all_score_fields[j]:
            scores = score_dfs[i,j][score_field].values
            print(score_field, np.mean(scores), np.std(scores))
```

```python
all_scores = list()
for i in range(files_arr.shape[0]):
    for j in range(files_arr.shape[1]):
        score_df = score_dfs[i,j]
        if j == 0:
            scores = score_df["rmsd_0"].values
        else:
            scores = score_df["rmsd_0+rmsd_1"].values

        all_scores.extend(scores)
```
