```python
import pickle 
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.lines import Line2D
import seaborn as sns
sns.set_theme()
import math
import pandas as pd
```

```python
stat_df_1 = pd.read_csv(Path(Path.home(), "Documents/xray/sample_bench/data/3ca7/78_sn_real_4x/score_analysis.csv"))
stat_df_2 = pd.read_csv(Path(Path.home(), "Documents/xray/sample_bench/data/3ca7/79_sn_real_nsb/score_analysis.csv"))
```

```python
color_on = True 

colors = list() 
alphas=list() 
edge_colors=list() 

colors_dict = {0: "#a6def2", 24: "#e5b3e4", 49: "#9eeea1"}

for i in range(50): 
    if i in colors_dict.keys() and color_on: 
        colors.append(colors_dict[i])
        alphas.append(1)
        edge_colors.append("black")
    else:
        colors.append("b")
        alphas.append(.3)
        edge_colors.append("none")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(30,6))
ax1,ax2 = axs[0],axs[1]

ax1.scatter(stat_df_1["min_xray_rmsd_avg"],stat_df_2["min_xray_rmsd_avg"], c=colors, alpha=alphas, s=100, edgecolors=edge_colors, marker="o")
ax2.scatter(stat_df_1["min_xray_r_free_0"],stat_df_2["min_xray_r_free_0"], c=colors, alpha=alphas, s=100, edgecolors=edge_colors, marker="o")

# ax1.set(xlim=(xmin, xmax))
# ax2.set(xlim=(0, .6))

plt.savefig(Path(Path.home(), "Desktop/natives_correlation_nsb.png"))
```
