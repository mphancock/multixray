```python
from pathlib import Path 
import pandas as pd 
import random
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_theme()
```

```python
job_num = 386258
xray_dir = Path(Path.home(), "Documents/xray")
log_stats_dir = Path(xray_dir, "sample_bench/out/3ca7/log_stats")
log_stats_file = Path(log_stats_dir, "{}.csv".format(job_num))

log_stats_df = pd.read_csv(log_stats_file, index_col=0)
log_stats_df.head()
```

```python
log_stats_df["std_dev_min_score"]
```

```python
plt.figure(figsize=(20,10))
# plt.scatter(log_stats_df["n"], log_stats_df["mean_min_score"])

plt.errorbar(log_stats_df["n"], log_stats_df["mean_min_score"], log_stats_df["std_dev_min_score"], linestyle='None', marker='o', capsize=3)
figure_file = Path(xray_dir, "sample_bench/out/3ca7/figures/{}.png".format(job_num))
plt.savefig(figure_file)

```
