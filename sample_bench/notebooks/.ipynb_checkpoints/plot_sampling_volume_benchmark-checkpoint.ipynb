```python
from pathlib import Path 
import pandas as pd 
import random
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_theme()
import matplotlib.colors as mcolors
from matplotlib.lines import Line2D
```

```python
job_nums = list() 
job_nums.append(788243)
job_nums.append(788244)
job_nums.append(788245)
job_nums.append(788246)

xray_dir = Path(Path.home(), "Documents/xray")
log_stats_dir = Path(xray_dir, "sample_bench/out/3ca7/single_md/log_stats")
fig_dir = Path(xray_dir, "sample_bench/figures")

log_stats_dfs = list() 
for job_num in job_nums:
    log_stats_file = Path(log_stats_dir, "{}.csv".format(job_num))
    log_stats_df = pd.read_csv(log_stats_file, index_col=0)
    log_stats_dfs.append(log_stats_df)
```

```python
if len(log_stats_dfs)%2 == 1: 
    n_row = len(log_stats_dfs)//2+1
else: 
    n_row = len(log_stats_dfs)//2
n_row
```

```python
fig, axs = plt.subplots(n_row, 2, figsize=(30,8*n_row))

colors = list() 
colors.append(mcolors.CSS4_COLORS["dodgerblue"])
colors.append(mcolors.CSS4_COLORS["orange"])
colors.append(mcolors.CSS4_COLORS["orangered"])

alpha = .5
space = 2

custom_legend = list() 
custom_legend.append(Line2D([], [], marker='.', markersize=10, color=colors[0], linestyle='None'))
custom_legend.append(Line2D([], [], marker='.', markersize=10, color=colors[1], linestyle='None'))

for i in range(len(log_stats_dfs)):
    log_stats_df = log_stats_dfs[i]
    ax1 = axs[i//2][i%2]
    ax1.errorbar(log_stats_df["n"][::space], log_stats_df["tot_mean"][::space], log_stats_df["tot_std"][::space], linestyle='None', marker='o', capsize=3, c=colors[0])
    if i == 0: 
        ax1.legend(custom_legend, ['score', 'rmsd'], loc="upper right", fontsize=15)
        
    ax2 = ax1.twinx()
    ax2.errorbar(log_stats_df["n"][::space], log_stats_df["rmsd_mean"][::space], log_stats_df["rmsd_std"][::space], linestyle='None', marker='o', capsize=3, c=colors[1], alpha=alpha)

# plt.show()
fig_file = Path()
plt.savefig(fig_file)

```
