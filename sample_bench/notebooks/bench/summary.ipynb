```python
from pathlib import Path
import pandas as pd
import numpy as np
from statannot import add_stat_annotation
```

```python
exp_dir = Path(Path.home(), "Documents/xray/sample_bench/data/7mhf/187_bench_ref_10000")
analysis_df = pd.read_csv(Path(exp_dir, "sample.csv"), index_col=0)
analysis_df.head()

```

```python
summary_df = pd.DataFrame()

for k in range(10):
    job_ids = [k*2, k*2+1, 20+k*2, 20+k*2+1]
    print(k)
    for N in [1, 2]:
        for J in [1, 2]:
            subset_df = analysis_df[(analysis_df["N"] == N) & (analysis_df["J"] == J) & (analysis_df["job_id"].isin(job_ids))]
            best_row = subset_df[subset_df["r_free_0"] == subset_df["r_free_0"].min()]
            r_free = best_row["r_free_0"].values[0]

            summary_df.loc[k, "r_free_{}_{}".format(N, J)] = r_free

            print(r_free)

# summary_df = pd.concat(best_rows)
summary_df.to_csv(Path(exp_dir, "summary.csv"))
summary_df.head()
```

```python
delta_0 = summary_df["r_free_2_1"] - summary_df["r_free_1_1"]
delta_1 = summary_df["r_free_1_2"] - summary_df["r_free_1_1"]
delta_2 = summary_df["r_free_2_2"] - summary_df["r_free_1_1"]
delta_3 = summary_df["r_free_2_2"] - summary_df["r_free_2_1"]

print(np.mean(delta_0), np.mean(delta_1), np.mean(delta_2), np.mean(delta_3))
print(np.std(delta_0), np.std(delta_1), np.std(delta_2), np.std(delta_3))
```
