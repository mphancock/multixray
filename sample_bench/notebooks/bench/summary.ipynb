```python
from pathlib import Path
import pandas as pd
import numpy as np
from statannot import add_stat_annotation
```

```python
exp_dir = Path(Path.home(), "Documents/xray/sample_bench/data/7mhf/182_bench")

analysis_df = pd.read_csv(Path(exp_dir, "ref_10000.csv"), index_col=0)
analysis_df.head()

```

```python
best_rows = list()
for N_id in list(range(2)):
    for job_id in list(range(20)):
        subset_df = analysis_df[(analysis_df["n"] == N_id) & (analysis_df["j"] == job_id) & (analysis_df["cif_name"] == 0)]
        best_row = subset_df[subset_df["r_free"] == subset_df["r_free"].min()]
        # print(N_id, job_id)

        best_rows.append(best_row)

summary_df = pd.concat(best_rows)
summary_df.to_csv(Path(exp_dir, "summary.csv"), index=False)
summary_df.head()
```

```python
delta_1s = list()
delta_2s = list()

for i in range(10):
    js = [i*2, i*2+1]

    row_n1_J1 = summary_df[(summary_df["n"] == 0) & (summary_df["j"] == js[0])]
    r_free_n1_J1 = row_n1_J1.iloc[0]["r_free"]
    row_n1_J2 = summary_df[(summary_df["n"] == 0) & (summary_df["j"] == js[1])]
    r_free_n1_J2 = row_n1_J2.iloc[0]["r_free"]
    row_n2_J1 = summary_df[(summary_df["n"] == 1) & (summary_df["j"] == js[0])]
    r_free_n2_J1 = row_n2_J1.iloc[0]["r_free"]
    row_n2_J2 = summary_df[(summary_df["n"] == 1) & (summary_df["j"] == js[1])]
    r_free_n2_J2 = row_n2_J2.iloc[0]["r_free"]

    delta_1 = r_free_n1_J2-r_free_n1_J1
    delta_2 = r_free_n2_J2-r_free_n1_J2

    delta_1s.append(delta_1)
    delta_2s.append(delta_2)

    print(i, js ,r_free_n1_J1, r_free_n1_J2, r_free_n2_J1, r_free_n2_J2)
    # print(delta_1, delta_2)

print(np.mean(delta_1s), np.mean(delta_2s))
print(np.min(delta_1s), np.min(delta_2s))
```
