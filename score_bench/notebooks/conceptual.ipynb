```python
import numpy as np
from pathlib import Path

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.ticker import FuncFormatter
import matplotlib.colors as mcolors
from matplotlib.lines import Line2D
from matplotlib import cm
from matplotlib.font_manager import FontProperties
import seaborn as sns
sns.set_theme()
plt.rcParams["font.family"] = "Times New Roman"
from matplotlib.ticker import MaxNLocator
```

```python
mean = 0
std_dev = 1

x_values = np.linspace(-10, 10, 400)
y_values = np.exp(-((x_values - mean) ** 2) / (2 * std_dev ** 2))

y_scatt = -y_values + np.random.normal(0, 0.025, size=x_values.shape)

y_native = np.min(-y_values)
y_scatt_filt = y_scatt[y_scatt >= y_native]
x_scatt_filt = x_values[y_scatt >= y_native]

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
plt.scatter(x_scatt_filt, y_scatt_filt, color='black', s=5)
plt.scatter(0, y_native, color="red", s=10)

ax.set_xlim(-5, 5)
ax.set_ylim(-1.1, .2)

ax.tick_params(left=False, labelleft=False)
ax.tick_params(bottom=False, labelbottom=False)

ax.xaxis.set_major_locator(MaxNLocator(nbins=5))

plt.savefig(Path(Path.home(), "Desktop/concept_0.png"))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

def negative_sigmoid(x):
    return -1 / (1 + np.exp(-x))

x = np.linspace(0, 10, 400)
y = negative_sigmoid(x)

y_max, y_min = y.max(), y.min()
print(y_min)

ax.plot(x, y, color="black")

ax.tick_params(left=False, labelleft=False)
ax.tick_params(bottom=False, labelbottom=False)

ax.set_xlim(-1, 11)
# ax.set_ylim(-4.5, 2.5)
ax.yaxis.set_major_locator(MaxNLocator(nbins=5))

plt.plot([-1,11], [-1, -1],color="red", linestyle='--', linewidth=0.5)

plt.savefig(Path(Path.home(), "Desktop/concept_4.png"))
```
