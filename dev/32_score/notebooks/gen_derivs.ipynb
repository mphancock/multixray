```python
pdb_file = Path(Path.home(), "Documents/xray/data/pdbs/3k0m/3k0m_refined.pdb")

# xray_structure = iotbx.pdb.input(str(pdb_file)).xray_structure_simple(
#     crystal_symmetry=crystal_symmetry
# )

m = IMP.Model()
hs = list()
pids = list()
n_state = 4
for i in range(n_state):
    h = IMP.atom.read_pdb(str(pdb_file), m, IMP.atom.AllPDBSelector())
    hs.append(h)
    pids.extend(IMP.atom.Selection(h).get_selected_particle_indexes())

for state in range(n_state):
    cnt = 0
    for pid in IMP.atom.get_leaves(h):
        print(state, cnt, m.get_particle_name(pid))

        cnt += 1
```

```python
for state in range(n_state):
    cnt = 0
    for pid in IMP.atom.Selection(hs[i]).get_selected_particle_indexes():
        print(state, cnt, m.get_particle_name(pid))

        cnt += 1
```

```python
crystal_symmetry = cctbx.xray.crystal.symmetry(
    unit_cell=(42.4190, 51.7410, 88.6390, 90, 90, 90),
    space_group_symbol="P 21 21 21"
)

xray_structure = get_xray_structure(
    hs=hs,
    occs=[.25]*n_state,
    pids=pids,
    crystal_symmetry=crystal_symmetry
)
```

```python
n_scatt = len(xray_structure.scatterers())
n_scatt_per_state = int(n_scatt/n_state)
print(n_scatt, n_scatt_per_state)
```

```python
for state in range(n_state):
    print(state)
    for j in range(10):
        xray_structure.scatterers()[state*n_scatt_per_state+j].show()
```

```python
cif_file = Path(Path.home(), "Documents/xray/data/cifs/3k0m/3k0m.cif")
f_obs = miller_ops.get_miller_array(
    f_obs_file=cif_file,
    label="_refln.intensity_meas"
)
f_obs = miller_ops.clean_miller_array(f_obs)

# Set flags from file.
status_array = miller_ops.get_miller_array(
    f_obs_file=cif_file,
    label="_refln.status"
)
flags = status_array.customized_copy(data=status_array.data()=="f")
f_obs, flags = f_obs.common_sets(other=flags)

# f_obs = f_obs.resolution_filter(d_max=10, d_min=5)
# flags = flags.resolution_filter(d_max=10, d_min=5)
f_obs.size()
```

```python
# flags_copy = flags.deep_copy()
# for j in range(flags_copy.data().size()):
#     flags_copy.data()[j] = False

# flags_copy.data()[i] = True

# f_obs_tmp = f_obs.select(flags_copy.data())
# flags_tmp = flags.select(flags_copy.data())
# flags_tmp.data()[0] = False

# print(f_obs_tmp.indices()[0], f_obs_tmp.data()[0])

xray_structure.scatterers().flags_set_grads(
    state=False
)
xray_structure.scatterers().flags_set_grad_site(
    iselection=xray_structure.all_selection().iselection()
)
xray_structure.scatterers().flags_set_grad_occupancy(
    iselection=xray_structure.all_selection().iselection()
)

f_model_manager = mmtbx.f_model.manager(
    xray_structure=xray_structure,
    f_obs=f_obs,
    r_free_flags=flags,
    target_name="ml"
)

fmodels = mmtbx.fmodels(fmodel_xray=f_model_manager)
fmodels.update_xray_structure(
    xray_structure=xray_structure,
    update_f_calc=True
)

# f_model_manager.update_all_scales(
#     apply_scale_k1_to_f_obs=True,
#     remove_outliers=False
# )

fmodels = mmtbx.fmodels(fmodel_xray=f_model_manager)
fmodels.update_xray_structure(
    xray_structure=xray_structure,
    update_f_calc=True
)

fmodels_target_and_gradients = fmodels.target_and_gradients(compute_gradients=True)
score = fmodels_target_and_gradients.target()
grads = fmodels_target_and_gradients.gradients()

print(f_obs.indices()[0], sum(grads))
```
