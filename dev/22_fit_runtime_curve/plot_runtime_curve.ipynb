```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define your three points (x, y)
xs = np.array([110, 100, 50, 60])
ys = np.array([269.99387073516846, 200.3917031288147, 27.349912405014038, 43.924341678619385])

# Fit a quadratic function (y = ax^2 + bx + c) to the data
def quad(x, a, c):
    return a * x**2 + c

params, covariance = curve_fit(quad,xs,ys,maxfev=10000)
a, c = params 
# Create a range of x values for the fitted curve
x_fit = np.linspace(min(xs), max(xs), 100)
y_fit = a * x_fit**2 + c

# Plot the original points and the fitted quadratic curve
plt.scatter(xs, ys, label='Data Points')
plt.plot(x_fit, y_fit, label='Quadratic Fit', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Quadratic Fit to Three Points')
plt.grid(True)
plt.show()
```
